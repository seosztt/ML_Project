## 각 변수와 'COST'와의 관계

### pairplot

```python
import seaborn as sns
sns.pairplot(X[X['PUM_NM']=='배추'])
```

```python
Napa=X[X['PUM_NM']=='배추']
```

```python
from sklearn.preprocessing import LabelEncoder

feature_LE=['WHSAL_NM','CMP_NM','PUM_NM','KIND_NM','POJ_NM','SIZE_NM','LV_NM','SAN_NM']
LE=LabelEncoder()

for i in feature_LE:
    temp1=LE.fit_transform(Napa[i])
    Napa[i]=temp1
```

```python
Napa[['LV_NM']].value_counts()
```

```python
Napa
```

```python
feature_LE.append('COST')
```

```python
Napa[feature_LE]
```

```python
feature_LE.remove('PUM_NM')
```

```python
sns.pairplot(Napa[feature_LE])
```

```python
sns.pairplot(Napa[(Napa['COST']<10000) & (Napa['COST']>0)][feature_LE])
```

### boxplot

```python
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='Malgun Gothic')
plt.rcParams['figure.figsize'] = (10, 6)
sns.boxplot(x='LV_NM',y='COST', data=X[(X['COST']<50000) & (X['COST'] > 0)])
plt.xticks(rotation=70)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (16, 6)
sns.boxplot(x='WHSAL_NM',y='COST', data=X[(X['COST']<50000) & (X['COST'] > 0)])
plt.xticks(rotation=70)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (24, 6)
sns.boxplot(x='CMP_NM',y='COST', data=X[(X['COST']<50000) & (X['COST'] > 0)])
plt.xticks(rotation=70)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (10, 6)
sns.boxplot(x='KIND_NM',y='COST', data=X[(X['PUM_NM']=='배추') & (X['COST']<50000) & (X['COST'] > 0)])
plt.xticks(rotation=70)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (10, 6)
sns.boxplot(x='POJ_NM',y='COST', data=X[(X['COST']<50000) & (X['COST'] > 0)])
plt.xticks(rotation=70)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (25, 6)
sns.boxplot(x='SIZE_NM',y='COST', data=X[(X['COST']<50000) & (X['COST'] > 0)])
plt.xticks(rotation=70)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (50, 16)
sns.boxplot(x='SAN_NM',y='COST', data=X[(X['PUM_NM']=='배추') & (X['COST']<50000) & (X['COST'] > 0)])
plt.xticks(rotation=70)
plt.show()
```
