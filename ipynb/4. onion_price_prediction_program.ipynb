### city

```python
LE_city=LabelEncoder()

CSP['city']=LE_city.fit_transform(CSP['city'])
```

```python
LE_city.classes_
```

```python
def get_integer_mapping(le):
    '''
    Return a dict mapping labels to their integer values
    from an SKlearn LabelEncoder
    le = a fitted SKlearn LabelEncoder
    '''
    res = {}
    for cl in le.classes_:
        res.update({cl:le.transform([cl])[0]})

    return res
```

```python
LE_city_mapping=get_integer_mapping(LE_city)
```

```python
kind_map=CSP.groupby(['city'])['kind'].agg(**{
    'kind':lambda x:x.mode()
}).reset_index()

market_map=CSP.groupby(['city'])['market'].agg(**{
    'market':lambda x:x.mode()
}).reset_index()

corp_map=CSP.groupby(['city'])['corp'].agg(**{
    'corp':lambda x:x.mode()
}).reset_index()

wide_map=CSP.groupby(['city'])['wide'].agg(**{
    'wide':lambda x:x.mode()
}).reset_index()
```
