# preprocess

```python
import pandas as pd

CSP=pd.read_csv('/content/drive/MyDrive/2조/datasets/전처리/PATIW.csv')
```

```python
CSP
```

## LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder

LE=LabelEncoder()
CSP['weather']=LE.fit_transform(CSP['weather'])
```

```python
CSP
```

## Split Train Test

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(CSP,
                               test_size = 0.2,
                               random_state = 2045)
```

```python
train=train.reset_index()
test=test.reset_index()
```

```python
train.drop(columns=['index'], inplace=True)
test.drop(columns=['index'], inplace=True)
```

```python
train
```

```python
test
```

## MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
MMS=MinMaxScaler(feature_range=(0,100))

MMS_list=['area','ratio','amount','inn','out']

train[MMS_list]=MMS.fit_transform(train[MMS_list])
test[MMS_list]=MMS.transform(test[MMS_list])
```

```python
test
```

# modeling

## Read

```python
import pandas as pd

train=pd.read_csv('/content/drive/MyDrive/2조/임시/datasets/전처리/train_PATIW_LE_MMS.csv')
test=pd.read_csv('/content/drive/MyDrive/2조/임시/datasets/전처리/test_PATIW_LE_MMS.csv')
```

```python
train.drop(columns=['Unnamed: 0'],inplace=True)
test.drop(columns=['Unnamed: 0'],inplace=True)
```

## RandomForest

### all columns

```python
X_train=train.drop(columns=['price'])
y_train=train['price']
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score

RF=RandomForestRegressor(n_estimators = 2000,
                            max_depth = 5,
                            criterion ='mse',
                            n_jobs = -1,
                            random_state = 2045)

scores=cross_val_score(RF,
                       X_train,
                       y_train,
                       scoring='neg_mean_squared_error',
                       cv=5)
```

```python
import numpy as np
np.sqrt(-scores.mean())
```
